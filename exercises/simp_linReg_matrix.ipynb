```python
import numpy as np
X=np.array([[1,1],[1,2],[1,3],[1,4],[1,5],[1,6]])
Y=np.array([[3],[5],[7],[9],[11],[13]])
beta=np.linalg.inv(X.T.dot(X)).dot(X.T).dot(Y)
print("Beta values: " + str(beta))
print("A0: " + str(beta[0][0]))
print("A1: " + str(beta[1][0]))
print("Predicted value for x=7: " + str(beta[0][0] + beta[1][0]*7))
```
