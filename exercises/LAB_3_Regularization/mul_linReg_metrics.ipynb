```python
import numpy as np
X=np.array([[1,2,60],[1,3,65],[1,4,70],[1,5,75],[1,6,80],[1,7,85]])
Y=np.array([[40],[48],[55],[63],[70],[78]])
beta=np.linalg.pinv(X.T@X)@X.T@Y
print("Beta values: " + str(beta))
print("A0: " + str(beta[0][0]))
print("A1: " + str(beta[1][0]))
print("A2: " + str(beta[2][0]))
print("Predicted value for x1=8 and x2=90: " + str(beta[0][0] + beta[1][0]*8 + beta[2][0]*90))
mae = np.mean((Y - X.dot(beta))**2)
print("Mean Squared Error: " + str(mae))
mse = np.mean((Y - X.dot(beta))**2)
print("Mean Squared Error: " + str(mse))
rmse = np.sqrt(mse)
print("Root Mean Squared Error: " + str(rmse))
r2 = 1 - (np.sum((Y - X.dot(beta))**2) / np.sum((Y - np.mean(Y))**2))
print("R-squared: " + str(r2))
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
X=np.array([[1,2,60],[1,3,65],[1,4,70],[1,5,75],[1,6,80],[1,7,85]])
Y=np.array([[40],[48],[55],[63],[70],[78]])
beta=np.linalg.pinv(X.T@X)@X.T@Y
print("Beta values: " + str(beta))
print("A0: " + str(beta[0][0]))
print("A1: " + str(beta[1][0]))
print("A2: " + str(beta[2][0]))
print("Predicted value for x1=8 and x2=90: " + str(beta[0][0] + beta[1][0]*8 + beta[2][0]*90))
mae = mean_absolute_error(Y, X.dot(beta))
print("Mean Absolute Error: " + str(mae))
mse = mean_squared_error(Y, X.dot(beta))
print("Mean Squared Error: " + str(mse))
rmse = np.sqrt(mse)
print("Root Mean Squared Error: " + str(rmse))
r2 = r2_score(Y, X.dot(beta))
print("R-squared: " + str(r2))
```
