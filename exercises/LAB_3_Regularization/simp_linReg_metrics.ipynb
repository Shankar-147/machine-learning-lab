```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
X=np.array([1,2,3,4,5,6])
Y=np.array([3,5,7,9,11,13])
mean_x = np.mean(X)
mean_y = np.mean(Y)
print("Mean of X: " + str(mean_x))
print("Mean of Y: " + str(mean_y))
a1 = np.sum((X-mean_x)*(Y-mean_y))/np.sum((X-mean_x)**2)
a0 = mean_y - a1*mean_x
print("A0: " + str(a0))
print("A1: " + str(a1))
print("Predicted value of Y for X=7: " + str(a0 + a1*7))
print("Mean Absolute Error: " + str(mean_absolute_error(Y, a0 + a1*X)))
print("Mean Squared Error: " + str(mean_squared_error(Y, a0 + a1*X)))
print("Root Mean Squared Error: " + str(np.sqrt(mean_squared_error(Y, a0 + a1*X))))
print("R-squared: " + str(r2_score(Y, a0 + a1*X)))
```
