```python
# ==========================================
# Part D : Ridge Regression
# ==========================================

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

# ------------------------------------------
# Load Dataset
# ------------------------------------------

df = pd.read_csv("../../datasets/house_regression_lab_dataset(in).csv")

# Independent Variables
X = df[["Area_sqft", "Bedrooms", "Age_years"]]

# Target Variable
y = df["Price_lakhs"]

# ------------------------------------------
# Split Dataset
# ------------------------------------------

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

# ==========================================
# Linear Regression Model
# ==========================================

linear_model = LinearRegression()

linear_model.fit(X_train, y_train)

linear_pred = linear_model.predict(X_test)

# ==========================================
# Ridge Regression Model
# ==========================================

ridge_model = Ridge(alpha=1.0)

ridge_model.fit(X_train, y_train)

ridge_pred = ridge_model.predict(X_test)

# ------------------------------------------
# 1. Ridge Intercept
# ------------------------------------------

print("Ridge Intercept (β0):")

print(ridge_model.intercept_)

# ------------------------------------------
# 2. Ridge Coefficients
# ------------------------------------------

print("\nRidge Coefficients")

print("Area_sqft :", ridge_model.coef_[0])

print("Bedrooms  :", ridge_model.coef_[1])

print("Age_years :", ridge_model.coef_[2])

# ------------------------------------------
# Ridge Regression Equation
# ------------------------------------------

print("\nRidge Regression Equation")

print(
    f"Price = {ridge_model.intercept_:.4f}"
    f" + ({ridge_model.coef_[0]:.4f} × Area_sqft)"
    f" + ({ridge_model.coef_[1]:.4f} × Bedrooms)"
    f" + ({ridge_model.coef_[2]:.4f} × Age_years)"
)

# ------------------------------------------
# 3. Evaluation Metrics
# ------------------------------------------

mae = mean_absolute_error(y_test, ridge_pred)

mse = mean_squared_error(y_test, ridge_pred)

rmse = np.sqrt(mse)

r2 = r2_score(y_test, ridge_pred)

print("\nEvaluation Metrics")

print("MAE  =", mae)

print("MSE  =", mse)

print("RMSE =", rmse)

print("R²   =", r2)

# ------------------------------------------
# 4. Compare Linear and Ridge Coefficients
# ------------------------------------------

comparison = pd.DataFrame({

    "Feature": X.columns,

    "Linear Regression": linear_model.coef_,

    "Ridge Regression": ridge_model.coef_

})

print("\nCoefficient Comparison")

print(comparison)

# ------------------------------------------
# Prediction Table
# ------------------------------------------

result = pd.DataFrame({

    "Actual Price": y_test.values,

    "Predicted Price": ridge_pred,

    "Error": y_test.values - ridge_pred

})

print("\nPrediction Table")

print(result)
```
