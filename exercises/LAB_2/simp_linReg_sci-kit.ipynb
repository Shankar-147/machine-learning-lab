```python
import numpy as np
from sklearn.linear_model import LinearRegression

X = np.array([[1],[2],[3],[4],[5],[6]])
Y = np.array([3, 5, 7, 9, 11, 13])

model = LinearRegression()
model.fit(X, Y)

a0 = model.intercept_
a1 = model.coef_[0]

print("Intercept (a0):", a0)
print("Coefficient (a1):", a1)

x1 = 7
y1 = a0 + a1 * x1

print("Predicted value for X =", x1, ":", y1)

x2 = 10
y2 = a0 + a1 * x2

print("Predicted value for X =", x2, ":", y2)
```
