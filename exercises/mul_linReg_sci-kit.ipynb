```python
import numpy as np
from sklearn.linear_model import LinearRegression

X = np.array([[2, 60],[3, 65],[4, 70],[5, 75],[6, 80],[7, 85]
])
Y = np.array([40, 48, 55, 63, 70, 78])

model = LinearRegression()

model.fit(X, Y)

a0 = model.intercept_
a1 = model.coef_[0]
a2 = model.coef_[1]

print("Intercept (A0):", a0)
print("Coefficient for Study Hours (A1):", a1)
print("Coefficient for Attendance (A2):", a2)

study_hours = 8
attendance = 90

predicted_marks = a0 + (a1 * study_hours) + (a2 * attendance)

print("\nPrediction")
print("Study Hours =", study_hours)
print("Attendance =", attendance)
print("Predicted Marks =", predicted_marks)
```
